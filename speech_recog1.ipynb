```python
from IPython.display import Audio
from google.colab import files
from pydub import AudioSegment
import io
import numpy as np  # Import numpy

# Upload your audio file to Colab
uploaded = files.upload()

# Read the audio file
for fn in uploaded.keys():
    # Load the Ogg file using pydub
    audio = AudioSegment.from_file(io.BytesIO(uploaded[fn]), format="ogg")

    # Convert the audio to WAV format
    wav_data = audio.raw_data
    sample_width = audio.sample_width
    sample_rate = audio.frame_rate
    audio_data = np.frombuffer(wav_data, dtype=np.int16)

# Play the audio
Audio(data=audio_data, rate=sample_rate)
```

```python
import speech_recognition as sr
from google.colab import files
!apt-get install ffmpeg -y

# Convert the audio file using FFmpeg



# Upload your audio file to Colab
uploaded = files.upload()


# Initialize the recognizer
recognizer = sr.Recognizer()

# Process each uploaded file
for fn in uploaded.keys():
    # Load the audio file
    audio_file = sr.AudioFile(io.BytesIO(uploaded[fn]))

    # Recognize the speech
    with audio_file as source:
        print("Processing audio file:", fn)
        audio_data = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio_data)
            print("Text from speech:", text)
        except sr.UnknownValueError:
            print("Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))
```
