```
import speech_recognition as sr
recognizer = sr.Recognizer()
```

```
def recognize_speech():
    with sr.Microphone() as source:
        print("Speak now:")
        recognizer.adjust_for_ambient_noise(source, duration=2)
        audio = recognizer.listen(source)
        try:
            print("Recognizing...")
            text = recognizer.recognize_google(audio)
            return text
        except sr.UnknownValueError:
            print("Sorry, could not understand the audio.")
            return None
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
            return None
```

```
from IPython.display import Audio
from google.colab import files
from pydub import AudioSegment
import io
import numpy as np
import speech_recognition as sr
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
from gtts import gTTS

# Step 1: Upload your audio file to Colab
uploaded = files.upload()

# Initialize the recognizer
recognizer = sr.Recognizer()

# Create variables to store recognized text and translation
recognized_text = ""
translation = ""

# Load the pre-trained translation model and tokenizer
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")

# Process each uploaded file
for fn in uploaded.keys():
    # Step 2: Check if the file is in WAV format, if not, convert it
    if not fn.endswith('.wav'):
        # Load the audio file using pydub
        audio = AudioSegment.from_file(io.BytesIO(uploaded[fn]), format="ogg")

        # Convert the audio to WAV format
        wav_data = audio.raw_data
        sample_width = audio.sample_width
        sample_rate = audio.frame_rate
        audio_data = np.frombuffer(wav_data, dtype=np.int16)

        # Save the converted WAV file
        wav_filename = fn.replace('.ogg', '_converted.wav')
        audio.export(wav_filename, format="wav")

        print(f"Converted WAV file saved as {wav_filename}")

        # Set the current file to the converted WAV file
        fn = wav_filename

    # Step 3: Load the audio file
    audio_file = sr.AudioFile(fn)

    # Recognize the speech
    with audio_file as source:
        print("Processing audio file:", fn)
        audio_data = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio_data)
            print("Text from speech:", text)
            recognized_text = text
        except sr.UnknownValueError:
            print("Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))

# Step 4: Tokenize the recognized text
model_inputs = tokenizer(recognized_text, return_tensors="pt")

# Step 5: Translate from English to Hindi
generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["ta_IN"]
)

# Step 6: Decode the generated tokens to get the translation
translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]

# Print the recognized speech and translation
print("Recognized Speech:", recognized_text)
print("Tamil Translation:", translation)

# Step 7: Convert the translated Tamil text to speech
tts = gTTS(text=translation, lang='ta')
tts.save('translated_tamil.mp3')

# Step 8: Play the audio
audio_path = 'translated_tamil.mp3'
Audio(audio_path, autoplay=True)
```
